```python
# load obwody 2025

file_path = os.path.join(POLAND_RAW_DATA, "obwody_2025.csv")

# Load the CSV with correct separator and encoding (probably cp1250 or utf-8-sig)
df = pd.read_csv(file_path, sep=";")

# Create the mapping: TERYT gminy â†’ Kod pocztowy
teryt_to_postal = df[["TERYT gminy", "Kod pocztowy"]].dropna().drop_duplicates()


teryt_to_postal["TERYT gminy"] = teryt_to_postal["TERYT gminy"].astype(int)

teryt_postal_map = dict(zip(teryt_to_postal["TERYT gminy"], teryt_to_postal["Kod pocztowy"]))

# Preview
print(teryt_postal_map)
```
